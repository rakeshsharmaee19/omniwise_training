```python
# pandas and pivot table
# Pivot Tables can give a snapshot view of the entire dataset which eases the process of inferring useful insights and 
#    making important observations in the dataset.
import pandas as pd
stocks = pd.read_csv('https://gist.githubusercontent.com/alexdebrie/b3f40efc3dd7664df5a20f5eee85e854/raw/ee3e6feccba2464cbbc2e185fb17961c53d2a7f5/stocks.csv')
print(stocks)

#we can use two methods pivot() and pivot_table()
```

```python
#pandas pivot syntax
# dataFrame.pivot(index='row values',columns='column values',values='data fields to view')
# dataFrame.pivot_table(data, values=None, index=None, columns=None, aggfunc=’mean’, 
#    fill_value=None, margins=False, dropna=True, margins_name=’All’, observed=False)
#labels 
#data=dataframe, the dataset whose pivot table is to be made
# values = colums, whose statistical summary is to be seen
# index = Column, rows value
# columns = Column value
```

```python
print(stocks)
stocks.pivot(index='symbol', columns=['date','open'],values='volume')
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'First Name': ['Aryan', 'Rohan', 'Riya', 'Yash', 'Siddhant', 'Mohan'],
                   'Last Name': ['Singh', 'Agarwal', 'Shah', 'Bhatia', 'Khanna','Singh'],
                   'Type': ['Full-time Employee', 'Intern', 'Full-time Employee', 
                            'Part-time Employee', 'Full-time Employee', 'Intern'],
                   'Department': ['Administration', 'Technical', 'Administration', 
                                  'Technical', 'Management','Management'],
                   'YoE': [2, 3, 5, 7, 6, 2],
                   'Salary': [20000, 5000, 10000, 10000, 20000,6000]})

print(df)
```

```python
output = pd.pivot_table(data=df, 
                        index='Type', 
                        columns='Department', 
                        values='Salary',
                       dropna=True)
print(output)
```

```python
output=pd.pivot(data=df,columns='Salary')
print(output)
```

```python
# pivot table   with multiple aggregate function
# if column parameter is not specified, it will aggregate based on the index
print(df)
output = pd.pivot_table(data=df, index='Type',
                        values='Salary',
                        aggfunc=['sum', 'mean', 'count'])
print(output)
output = pd.pivot_table(data=df, index='Type',
                        columns='Department',
                        values='Salary',
                        aggfunc=['sum', 'mean', 'count'])
print(output)
```

```python
# calculating sum of the colums
print(df)
output = pd.pivot_table(data=df, index=['Type'],
                        values='Salary',
                        aggfunc=['sum', 'mean', 'count'],
                        margins=True,
                        margins_name='Grand Total')
print(output)
```

```python
# aggregate for multiple feature
# here total aggregate and the value should be same
print(df)
output = pd.pivot_table(data=df, index='Type', values=['Salary', 'YoE'],
                        columns=['Department'],
                        aggfunc={'Salary': np.sum, 'YoE': np.mean},
                       margins=True,
                       margins_name='Grand Total')
print(output)
```

```python
#we can also fill the value if not available
print(df)
output = pd.pivot_table(data=df, index='Type', values=['Salary', 'YoE'],
                        columns=['Department'],
                        aggfunc={'Salary': np.sum, 'YoE': np.mean},
                        fill_value='Not applicable')

print(output)
```

```python
# multilabeled index
print(df)
output = pd.pivot_table(data=df, index=['Type', 'First Name'],
                        values='Salary',
                        aggfunc=['sum', 'mean', 'count'],
                        margins=True,
                        margins_name='Grand Total')
print(output)
```
